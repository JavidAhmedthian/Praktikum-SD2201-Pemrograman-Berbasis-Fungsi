```python
# Decorator untuk autentikasi
def require_login(func):
    def wrapper(*args, **kwargs):
        # Cek apakah user sudah login
        if not args[0].is_authenticated:  # Misalkan is_authenticated adalah atribut yang menandakan apakah user sudah login atau tidak
            print("Anda harus login untuk mengakses fitur ini!")
            return
        return func(*args, **kwargs)
    return wrapper

# Kelas Aplikasi Pengelolaan Keuangan Sederhana
class Keuanganku:
    def __init__(self):
        self.is_authenticated = False
        self.saldo = 0

    @require_login
    def tambah_transaksi(self, jumlah):
        self.saldo += jumlah
        print(f"Transaksi berhasil! Saldo sekarang: {self.saldo}")

    def login(self, username, password):

        if username == "user" and password == "123":
            self.is_authenticated = True
            print("Login berhasil!")
        else:
            print("Username atau password salah!")

    def logout(self):
        self.is_authenticated = False
        print("Logout berhasil.")

# Fungsi untuk mendapatkan input dari pengguna
def main():
    app = Keuanganku()

    while True:
        print("\nMenu:")
        print("1. Login")
        print("2. Logout")
        print("3. Tambah Transaksi")
        print("4. Keluar")

        pilihan = input("Pilih menu: ")

        if pilihan == "1":
            username = input("Masukkan username: ")
            password = input("Masukkan password: ")
            app.login(username, password)
        elif pilihan == "2":
            app.logout()
        elif pilihan == "3":
            jumlah = float(input("Masukkan jumlah transaksi: "))
            app.tambah_transaksi(jumlah)
        elif pilihan == "4":
            break
        else:
            print("Pilihan tidak valid!")

if __name__ == "__main__":
    main()
```
